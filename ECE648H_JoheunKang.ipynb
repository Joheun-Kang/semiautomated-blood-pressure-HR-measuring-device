## Apply bandpass filter

```python
# apply bandpass filter

from scipy import signal 
import numpy as np
fs = 1/0.1      #freq = ~10ms
fc = np.array([0.9,1.7])
wc = 2*fc/fs
offset2 = data
b,a = signal.butter(5,wc,btype= "bandpass", analog = False)
filtered_data = signal.filtfilt(b,a,offset2,axis = 0)                                                                                                                                                                                                                       
```

```python
relief_point =np.argmax(data) # 
print('you relief the valve at sample number :%d ' % relief_point )

plt.title("banpass filter applied version")
plt.xlabel("samples")
plt.ylabel("pressure deviation - mmHg")
plt.plot(range(len(filtered_data)),filtered_data)
```

```python
offset = data[relief_point:]

b,a = signal.butter(5,wc,btype= "bandpass", analog = False)
filtered_data2 = signal.filtfilt(b,a,offset,axis = 0)

plt.title("banpass filter applied version (Data from the valve relief point)")
plt.xlabel("samples")
plt.ylabel("pressure deviation - mmHg")
plt.plot(range(len(filtered_data2)),filtered_data2)
```

## Heart Rate from oscillation

```python
hr_data = list(filtered_data2.copy())


# plot amplitude graph from filtered_data2
amp_filtered = []
for i in range(1,len(hr_data)):
    amp_filtered.append(hr_data[i]-hr_data[i-1])

plt.title("amplitude of pulstation (from filtered_data2 )")
plt.xlabel("samples")
plt.ylabel("amplitude of pulstation ")
plt.plot(range(len(amp_filtered)),amp_filtered)


index2 = amp_filtered.index(max(amp_filtered))

prev_data = amp_filtered[:index2]

for i in range(len(prev_data)-2,-1,-1):
    # find first local maxima 
    if prev_data[i-1]<prev_data[i]>prev_data[i+1] and prev_data[i]>0:
        index1 = i # index for loca maxima
        break

print('index1',index1)
print('index2',index2)

distace = index2 - index1 # distance between local max
time_take = 10*distace # 10ms * distance 
print('time_take %3.2f ms' % time_take)
print('time_take %3.2f s ' % (time_take/100))


print("Your Heart Rate is %3.2f BPM" % (60/(time_take/100)))
```
